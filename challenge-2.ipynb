```python
def get_coordinates (where):

    url_geocode = f"https://geocode.xyz/{where}?json=1"
    
    try:
        res = requests.get(url_geocode).json()
        return res["longt"], res["latt"]
    
    except:
        print(f"Sorry, no matches for {where}")
        time.sleep(3)
        get_coordinates(where)
        
```

# 4. Define a function that returns name, lat, long for starbucks OR restaurants AND for any given location

```python
def foursquare_places (venue,location):
    url = f"https://api.foursquare.com/v3/places/search?query={venue}&near={location}"

    headers = {
        "accept": "application/json",
        "Authorization": token
    }

    response = requests.get(url, headers=headers).json()
    return response
```

```python
response= foursquare_places("starbucks","Ironhack Barcelona",1)
response2= foursquare_places("restaurants","Ironhack Barcelona",1)
```

```python
response2
```

```python
def everything_from_foursquare3 (one_element):
    
    address = one_element["location"]["formatted_address"]
    name = one_element["name"]
    lat = one_element["geocodes"]["main"]["latitude"]
    lon = one_element["geocodes"]["main"]["longitude"]
    dict3 = {"name": name, "lat": lat, "lon": lon}
    
    return dict3
```

```python
new_list = []
for i in response["results"]:
    new_list.append(everything_from_foursquare3 (i))
for x in response2["results"]:
    new_list.append(everything_from_foursquare3 (x))
df3 = pd.DataFrame(new_list)
df3
```
